```python
import pandas as pd 
```

```python
#Load dataset & view records
```

```python
df = pd.read_csv(r"D:\Download\StudentPerformanceFactors.csv")

print("Dataset Shape (Rows, Columns):")
print(df.shape)

print("\nFirst 5 records:")
print(df.head())

print("\nLast 5 records:")
print(df.tail())
```

```python
#  Manually identify feature types (by inspection)
```

```python
numerical_features = [
    "Hours_Studied", "Attendance", "Sleep_Hours",
    "Previous_Scores", "Tutoring_Sessions", "Exam_Score"
]

binary_features = [
    "Extracurricular_Activities", "Internet_Access", "Learning_Disabilities"
]

ordinal_features = [
    "Parental_Involvement", "Access_to_Resources", "Motivation_Level",
    "Family_Income", "Teacher_Quality",
    "Parental_Education_Level", "Distance_from_Home"
]

categorical_features = [
    "Gender", "School_Type", "Peer_Influence"
]

print("\nFeature Classification:")
print("Numerical Features:", numerical_features)
print("Binary Features:", binary_features)
print("Ordinal Features:", ordinal_features)
print("Categorical Features:", categorical_features)
```

```python
# df.info() & df.describe()
```

```python
print("\nDataset Info:")
df.info()

print("\nStatistical Summary:")
print(df.describe())


# ==================================================
# 4. Unique values in categorical columns
# ==================================================
print("\nUnique Values in Categorical / Ordinal / Binary Columns:")
for col in categorical_features + ordinal_features + binary_features:
    print(f"\n{col}:")
    print(df[col].unique())
```

```python
# Target variable & input features
```

```python
target_variable = "Exam_Score"
input_features = df.drop(columns=[target_variable]).columns.tolist()

print("\nTarget Variable:")
print(target_variable)

print("\nInput Features:")
print(input_features)
```

```python
# Dataset size & ML suitability
```

```python
rows, cols = df.shape

print("\nDataset Size Analysis:")
print(f"Number of Rows: {rows}")
print(f"Number of Columns: {cols}")

if rows > 1000:
    print("Dataset is suitable for Machine Learning.")
else:
    print("Dataset may be too small for ML.")
```

```python
# Data quality checks
```

```python
print("\nMissing Values per Column:")
print(df.isnull().sum())

print("\nBinary Feature Distribution (Imbalance Check):")
for col in binary_features:
    print(f"\n{col} value counts:")
    print(df[col].value_counts())
```
